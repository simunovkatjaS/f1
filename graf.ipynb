```python
import fastf1
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

import os
cache_path = '/content/drive/MyDrive/f1_cache'
os.makedirs(cache_path, exist_ok=True)

fastf1.Cache.enable_cache(cache_path)
```

```python
session = fastf1.get_session(2019,'Great Britain', 'R')
session.load()

fast_lap = session.laps.pick_fastest()
tel = fast_lap.get_car_data().add_distance()

plt.plot(tel['Distance'], tel['Speed'])
plt.xlabel('Distance (m)')
plt.ylabel('Brzina (km/h)')
plt.title(f"Najbrzi krug {fast_lap['Driver']}")
plt.show()
```
